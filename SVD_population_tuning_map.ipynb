```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from sklearn.metrics import pairwise_distances
from sklearn.preprocessing import StandardScaler
from collections import Counter
import pandas as pd
import scipy.stats as ss

z_thresh = 2
```

```python
def get_active_cells(traces):

    # going to return a dictionary with only active cells, formatted exactly the same as traces

    d = dict.fromkeys(traces.keys())

    for cell in traces:
        if traces[cell]['active'] == True:
            d[cell] = traces[cell]
        else:
            d.pop(cell,None)

    return d
```

```python
# load the first dictionary file
cell = "C:/Users/Conor/Documents/Imaging_Data/Two-Photon/Psilocybin_Project/Evoked_Cohort_Mice/ID186/12012022_GCaMP6s_ID186_psilo/TSeries-12012022-1321-027/suite2p/plane0/cells.pkl"

with open(cell,'rb') as f:
    cell_dict_1 = pickle.load(f)
```

```python
cell_dict_1 = get_active_cells(cell_dict_1)
cell_list = list(cell_dict_1.keys())
```

```python
nfreq = list(cell_dict_1[cell_list[0]]['traces'].keys())
nInt = list(cell_dict_1[cell_list[0]]['traces'][nfreq[0]].keys())
ntrial = list(cell_dict_1[cell_list[0]]['traces'][nfreq[0]][nInt[0]].keys())
print(nfreq, nInt, ntrial)
```

```python
tuning_array = np.array(cell_dict_1[cell_list[0]]['tuning'])
print(tuning_array)
```

```python
matrix_all = np.array([[[[cell_dict_1[n]['traces'][i][j][k] for n in cell_dict_1.keys()] for k in ntrial] for j in nInt] for i in nfreq])
print(matrix_all.shape)
```

```python
matrix_av = matrix_all.mean(axis=2)
print(matrix_av.shape)
matrix_res = matrix_av[:,:,:,5:].mean(axis=-1) - matrix_av[:,:,:,:5].mean(axis=-1)
print(matrix_res.shape)
```

```python
# Run single value decomposition on the population. a is frequency tuning, b is intensity tuning. S is the amount of variation explained by 
# freq and intensity
tuning = []
for i in range(matrix_res.shape[-1]):
    U, S, Vt = np.linalg.svd(matrix_res[:,:,i])
    a, b = np.argmax(np.abs(U[:,0])), np.argmax(np.abs(Vt[0]))
    S = S**2/(S**2).sum()
    if S[0] > 0.4: tuning.append([a,b])
```

```python
tuning = np.array(tuning)

print(tuning[0])
print(len(tuning))
```

```python
bins = np.arange(0,12,1)
plt.hist(tuning[0:,0],bins=bins,density=True)
```

```python
# Initialize a dict with keys for all possible frequencies and default value zero. 
counts = {key: 0 for key in (range(len(nfreq)))}
best_freqs = list(tuning[:,0])
# Iterate through BF list and update corresponding dict key if it appears as a BF. 
for i in range(len(best_freqs)):
    if best_freqs[i] in counts.keys():
        counts[best_freqs[i]] += 1

# Normalize the count to the total number of cells
for key,val in counts.items():
    counts[key] = val/len(cell_list)

plt.bar(range(len(counts)), counts.values(), align='center')
plt.xticks(range(len(nfreq)), nfreq)
plt.xticks(rotation=90)
plt.title("Distribution of Best Frequencies SVD Method")
plt.xlabel('Frequency Hz')
plt.ylabel('Normalized Count')
plt.show()
```
